```python
import tkinter as tk
root= tk.Tk()
root.title('Counter')

counter=0

def time_count(label):
    counter=0
    def count():
        global counter
        counter+=1
        label.config(text= str(counter))
        label.after(10,count)
    count()


file1=Label(root,fg='red',width=20,font=20)
file1.pack()

time_count(file1)
label1=Button(root,text='Stop Count',fg='red',width=10,font=10,command=root.destroy).pack()
root.mainloop()
```

```python
import tkinter as tk
root=tk.Tk()
root.title('Conter')

count_sec=0
count_min=0

def count_time(label):
    count_sec=0
    count_min=0
    def count():
        global count_sec
        global count_min
        count_sec+=1

        if count_sec==60:
            count_sec=0
            count_min+=1

        if count_sec<9:
            label.config(text='0'+str(count_min)+':'+'0'+str(count_sec))
        elif count_sec>=9 and count_min<=9:
            label.config(text='0'+str(count_min)+':'+str(count_sec))
        else:
            label.config(text=str(count_min)+':'+str(count_sec))

        label.after(1000,count)
    count()



label1=tk.Label(root,fg='purple',width=20,font=20)
label1.pack()
count_time(label1)

button=tk.Button(root,text='Stop Count',fg='pink',width=10,font=10,command=root.destroy).pack()
root.mainloop()

```

```python
import tkinter as tk 
root=tk.Tk()
root.title('Time Counter')
count_sec=0
count_min=0
def time(label1):
    count_sec=0
    count_min=0
    def count():
        global count_min
        global count_sec
        count_sec+=1
        if count_sec==60:
            count_sec=0
            count_min+=1
        if count_sec<=9:
            label1.config(text='0'+str(count_min)+":"+'0'+str(count_sec))
        elif count_sec>=9 and count_min<=9:
            label1.config(text='0'+str(count_min)+':'+str(count_sec))
        else :
            label1.config(text=str(count_min)+':'+str(count_sec))
        label1.after(10,count)
    count()

label=tk.Label(root,bg='purple',fg='white',font=20,width=50)
label.pack()
time(label)
button1=tk.Button(root,text='Stop Watch',bg='orange',fg='white',width=30,font=10,command=root.destroy).pack()
root.mainloop()
```

```python
from tkinter import *

canvas_width=800
canvas_height=600
python_green='red'

def polygon_star(canvas,x,y,p,t,outline='blue',fill='red',width=1):
    box=[]
    for i in (1,-1):
        box.extend((x,y+i*p))
        box.extend((x+i*t,y+i*t))
        box.extend((x+i*p,y))
        box.extend((x+i*t,y-i*t))

    print(box)

    canvas.create_polygon(box,outline=outline,fill=fill,width=width)
       

root=Tk()

w=Canvas(root,width=canvas_width,height=canvas_height)
w.pack()

p=45
t=5
step=8

step_x=int(canvas_width/step)
step_y=int(canvas_height/step)

for i in range(1, step):
    polygon_star(w,i*step_x,i*step_y,p,t,outline='yellow',fill='purple',width=3)

    polygon_star(w,i*step_x,canvas_height - i*step_y,p,t,outline='yellow',fill='purple',width=3)

root.mainloop()



```

```python
from pygame import mixer

#Instantiate mixer
mixer.init()

#Load audio file
mixer.music.load('water.mp3')

print('Music Started Playing...')

#Set preferred volume
mixer.music.set_volume(0.2)

#Play the music
mixer.music.play()

#Infintie loop
while True:
    print('*'*60)
    print("Press 'p' for pause the music\n")
    print("Press 'r' for resume the music\n")
    print("Press 'e' for exit the program")

    #take the user input
    userinput = input(' ')

    if userinput == 'p':
        
        #pause the music
        mixer.music.pause()
        print('Music is paused...')

    elif userinput == 'r':

        #resume the music
        mixer.music.unpause()
        print('Music is resumed...')


   
    elif userinput == 'e':

        #stop  the musicplayback
        mixer.music.stop()
        print('Music is stopped...')
        break
```
